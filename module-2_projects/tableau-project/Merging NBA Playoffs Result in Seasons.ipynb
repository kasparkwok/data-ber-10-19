## For NBA Salary (only from 03 to 17) Data Cleaning

```python
Salary = pd.read_csv("NBASalaryData03-17.csv")
```

```python
Salary.head()
```

```python
# Replace LA Clippers to Los Angeles Clippers and weird team names
Salary["Team"] =Salary["Team"].replace("LA Clippers","Los Angeles Clippers")
Salary["Team"] =Salary["Team"].replace("NO/Oklahoma City\r\r\n Hornets","New Orleans/Oklahoma City Hornets") 
Salary["Team"] =Salary["Team"].replace("NO/Oklahoma City Hornets","New Orleans/Oklahoma City Hornets") 
```

```python

# Convert Season to Year

# Convert season into year using regex 
Salary['Season'] = Salary['Season'].str.replace(r'(\d{2}-\d{2})', '').astype('int64')
Salary["Season"] =Salary["Season"].replace(1900,2000)
```

```python
#Add playoffs result to the salary chart

for i in range(1981, 2019):
    for j in list(NBA_Team_Full_Name.values()):
        if len(list(NBA_team_season[(NBA_team_season["Year"]==i) & (NBA_team_season["Full_Name"]==j)]["Playoffs"]))>0:
            Salary.loc[(Salary["Season"]==i) & (Salary["Team"] == j), "Playoffs"] = list(NBA_team_season.loc[(NBA_team_season["Year"]==i) & (NBA_team_season["Full_Name"]==j),"Playoffs"])[0] 
            
Salary["Playoffs"] = Salary["Playoffs"].fillna("Did not make the Playoffs")   
```

```python
#Add Team winning percentage

for i in range(1981, 2019):
    for j in list(NBA_Team_Full_Name.values()):
        if len(list(NBA_team_season[(NBA_team_season["Year"]==i) & (NBA_team_season["Full_Name"]==j)]["W/L%"]))>0:
            Salary.loc[(Salary["Season"]==i) & (Salary["Team"] == j), "Team Winning Percentage"] = list(NBA_team_season.loc[(NBA_team_season["Year"]==i) & (NBA_team_season["Full_Name"]==j),"W/L%"])[0] 
```
