## 2. Replacing NaN value to some meaningful values 

```python
database.isnull().sum()
```

After initial column cleaning, we should be proceeding to deal with the NaN values in the remaining columns.

```python
#replace NaN with "N/A" in "Country","Area","Location" columns if all of them return NaN
database.update(database[["Country","Area","Location"]][database["Country"].isnull() & database["Area"].isnull() & database["Location"].isnull()].fillna("N/A"))
```

```python
#replace NaN with "N/A" for "Area" and "Location"
database["Area"][database["Area"].isnull()]="N/A"
database["Location"][database["Location"].isnull()]="N/A"
```

```python
#check the remaining NaN values in "Country"
database[database["Country"].isnull()][["Country","Area","Location"]]
```

```python
database["Country"].value_counts()
```

```python
#Replacing them either with the corresponding values based on the given information or "international sea"

database.at[3792,"Country"]="Australia"
database.at[4412,"Country"]="Hong Kong"
database.at[4790,"Country"]="France"
database.at[5560,"Country"]="USA"
database.at[5847,"Country"]="Barbados"
database.at[5896,"Country"]="Hong Kong"


database["Country"][database["Country"].isnull()]="International Sea"
```
