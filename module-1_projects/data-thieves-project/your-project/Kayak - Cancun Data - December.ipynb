```python
#Create a for loop to generate the cheapest hotels for 3.0 4.0 and 5.0 stars

stars_rating= [3,4,5]
cheapest_hotel_list=pd.DataFrame(columns= Column_Names)
for i in stars_rating:
    list_with_star= Cancun_dec[Cancun_dec["Stars"]==i]
    cheapest_hotel= list_with_star[list_with_star["Price"] == list_with_star["Price"].min()]
    cheapest_hotel_list= cheapest_hotel_list.append(cheapest_hotel)

cheapest_hotel_list

cheapest_hotel_list["id"] = cheapest_hotel_list["City"]+"_"+"2019-12-21"
```
