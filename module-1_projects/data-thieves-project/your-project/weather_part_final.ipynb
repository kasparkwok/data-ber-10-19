Importing packages

```python
import pandas as pd
import requests
import json
import datetime
import matplotlib
%matplotlib inline
```

```python
dict_lat_long = {'Buenos Aires': ['-34.61315', '-58.37723', '-03:00'],
                 'Cancun': ['21.1742897', '-86.8465576', '-05:00'],
                 'Rio de Janeiro': ['-22.90278', '-43.2075', '-03:00']}
```

```python
def weather_dep():

    # Ask for inputs
    
    city = str(input('What is you destination:'))
    departure_date = str(input('When do you want to go (YYYY-MM-DD):'))
    days = int(input('How long do you want to stay:'))    
    
    # Create a list of dates to search for weather info
    
    years_hist = 3
    departure_list = departure_date.split('-')
    departure_day = datetime.date(year=int(departure_list[0]), month=int(departure_list[1]), day=int(departure_list[2]))
    date_list = [str(departure_day.replace(year=int(departure_list[0])-(y+1)) + datetime.timedelta(days=x)) for x in range(days) for y in range(years_hist)]
    
    # Create an empty DataFrame with the desired columns
    
    column_name = ['id','city','departure date', 'lower temperature', 'higher temperature', 'rain probability', 'weather condition']
    df_weather = pd.DataFrame(columns= column_name)
    
    # For loop to append the daily weather to the empty DataFrame
#    for c in city:        
    for period in date_list:
        url_base = f'https://api.darksky.net/forecast/5c060fde06e54a7eb56d4fbbce7dbd8f/{dict_lat_long[city][0]},{dict_lat_long[city][1]},{period}T00:00:00{dict_lat_long[city][2]}?units=ca&?exclude=currently,flags'
        headers = {'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                   'Accept-Encoding': 'gzip, deflate, br',
                   'Accept-Language': 'de,en-US;q=0.7,en;q=0.3',
                   'Connection': 'keep-alive',
                   'Host': 'api.darksky.net',
                   'Referer': 'https://darksky.net/dev/account',
                   'TE': 'Trailers',
                   'Upgrade-Insecure-Requests': '1',
                   'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:70.0) Gecko/20100101 Firefox/70.0'}
        r = requests.get(url_base, headers)
        row_list = pd.Series([str(city + '_' + departure_date), city, departure_date, r.json()['daily']['data'][0]['temperatureLow'], r.json()['daily']['data'][0]['temperatureHigh'], r.json()['daily']['data'][0]['precipProbability'],r.json()['daily']['data'][0]['summary']], index=column_name)
        df_weather = df_weather.append(row_list, ignore_index=True)
        row_list = row_list.drop(column_name)
    return df_weather        
```

```python
destination_summary = df_destination_total.groupby(by='id').mean().reset_index()
destination_summary['rain probability'] = destination_summary['rain probability']*100
destination_summary
```
