#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
fig, axes = plt.subplots(1, 3 , tight_layout=True)
hist_colour = ["brown", "green", "blue"]

no_of_bins = [10,20,50]
for b in range(len(no_of_bins)):
    titanic["Age"].hist(bins = no_of_bins[b], ax = axes[b], color = hist_colour[b])
```

#### Use seaborn to show the distribution of the column `Age`

```python
# your code here
fig, axes = plt.subplots(1, 3 , tight_layout=True, figsize = (30,15))
hist_colour = ["brown", "green", "blue"]
no_of_bins = [10,20,50]
for b in range(len(no_of_bins)):
    sns.distplot(titanic["Age"],bins = no_of_bins[b], ax = axes[b], color = hist_colour[b]);
```
